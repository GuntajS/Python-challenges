<h2>Watermarking and Image

```python
from PIL import Image
```

```python
mac = Image.open('example.jpg')
watermark = Image.open('_watermark.png')
```

```python
watermark.size
```

```python
mac.size
```

```python
new_watermark = watermark.crop((200,200,400,300))
```

```python
new_watermark = new_watermark.resize(((1400, 1200)))
```

```python
new_watermark.putalpha(100)
```

```python
new_watermark
```

```python
mac.paste(im=new_watermark,box=(0,0),mask=new_watermark)
```

```python
import sys # to access the system
import cv2
img = cv2.imread("example.jpg", cv2.IMREAD_ANYCOLOR)
 
while True:
    cv2.imshow("mac", img)
    cv2.waitKey(0)
    sys.exit() # to exit from all the processes
 
cv2.destroyAllWindows() # destroy all windows
```
