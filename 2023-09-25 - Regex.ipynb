<h2>Using The Pattern to find a Match</h2>

```python
text =  "My phone number is 123-999-4356"
```

```python
phone =  re.search(r'\d\d\d-\d\d\d-\d\d\d\d',text)
```

```python
phone
```

```python
phone.group()
```

```python
phone = re.search(r'\d{3}-\d{3}-\d{4}',text)
```

```python
phone
```

```python
phone_pattern = re.compile(r'(\d{3})-(\d{3})-(\d{4})')
```

```python
results = re.search(phone_pattern,text)
```

```python
results.group()
```

```python
results.group(1)
```

```python
results.group(2)
```

```python
results.group(5)
```

<h2>Additional Regex Syntax</h2>

```python
re.search(r'cat|dog','The cat is here')
```

```python
re.findall(r'.at', 'The cat in the hat sat there')
```

```python
re.findall(r'...at', 'The cat in the hat splat there')
```

```python
re.findall(r'^\d', 'the 2 is a number')
```

```python
re.findall(r'^\d', '2 is a number')
```

```python
re.findall(r'\d$', 'the 2 is a number 2')
```

```python
re.findall(r'\d$', 'the 2 is a number two')
```

```python
phrase = 'there are 3 numbers 34 inside 5 this sentence'
```

```python
pattern = r'[^\d]+'
re.findall(pattern,phrase)
```

```python
test_phrase = "This is a string! But it has punctuation. How can we remove it?"
```

```python
clean = re.findall(r'[^!.?]+',test_phrase) 
# after the ^ there are the punctuation to be removed
```

```python
' '.join(clean)
```

```python
text = 'Only find the hypen-words in this sentence. But you do not know how long-ish they are'
```

```python
patter = r'[\w]+-[\w]+' # could also be r'\w+-\w+'
```

```python
re.findall(patter,text)
```
